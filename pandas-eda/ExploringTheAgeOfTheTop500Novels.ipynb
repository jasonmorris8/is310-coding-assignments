```python
import altair as alt
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("novels.csv")
```

# Exploring the Age of the Top Ranked Novels #

```python
# Plotting the age of the novels using matplotlib
plt.figure(figsize=(10, 6))
plt.hist(df['pub_year'], bins=20, edgecolor='black')
plt.xlabel('Publication Year')
plt.ylabel('Count')
plt.title('Distribution of Publication Year')
plt.show()
```

This graph shows that the majority of the top 500 novels came from the year 1800 and later, with a large amount of them coming from the late 1900's.

## The age of the top 50 novels ##

```python
import altair as alt

# Select the top 50 novels
top_50_df = df.head(50)

# Create the Altair chart
chart = alt.Chart(top_50_df).mark_bar().encode(
).properties(
    title='Age of the Top 50 Novels',
    width=600,
    height=400
)
# Calculate the age of the novels
top_50_df['age'] = 2024 - top_50_df['pub_year']

# Update the encoding of the chart
chart = chart.encode(
    x='title',
    y='age'
)

# Show the updated chart
chart.show()
chart = chart.configure_axisX(
    labelAngle=45,
    labelFontSize=12
)
```

This graph displays the top 50 novels on the x axis and the y axis is the age of the novels.
It shows that there a only three novels in the top 50 that are over 250 years old, only 2 being over 300 years old, and only one being over 400 years old. 

## The newest and the oldest novel in the dataset ##

```python
oldest_novel = df[df['pub_year'] == df['pub_year'].min()]
newest_novel = df[df['pub_year'] == df['pub_year'].max()]

print("Oldest Novel:")
print(oldest_novel[['title', 'author', 'pub_year']])

print("\nNewest Novel:")
print(newest_novel[['title', 'author', 'pub_year']])
```

## Average age of the top 500 novels ##

```python
average_age =2024-df['pub_year'].mean()
print("Average Age of the Top 500 Novels:", average_age)
```
